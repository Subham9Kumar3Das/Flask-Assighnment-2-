### Q1. Explain GET and POST methods.

* GET is used to retrieve data from the server, and the data is visible in the URL.
* POST is used to send data to the server for processing, and the data is not visible in the URL.

### Q2. Why is request used in Flask?

The "request" object in Flask is used to get information from users who visit your website, like form data or input, so your Flask app can respond accordingly, making interactive web applications possible. It allows your Flask app to handle user input and send appropriate responses.

### Q3. Why is redirect() used in Flask?

In Flask, the "redirect()" function is used to send the user's web browser to a different URL. It allows your Flask application to respond to a request by telling the browser to go to another route or page. This is commonly used after processing a form submission or when you want to direct users to a different part of your website, making navigation smoother and more user-friendly.

### Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are files that define the structure of web pages with placeholders for dynamic content. The "render_template()" function is used to combine these templates with data from Python code, generating dynamic web pages for users.
